```python
import librosa
import numpy
import matplotlib.pyplot as plt
```

```python
audio,sfreq = librosa.load('songs/wav/first.wav')
```

```python
time = numpy.arange(0,len(audio))/sfreq
```

```python
time
plt.plot(audio[0:1024])
```

```python
CHUNK = 1024
start = 0
end = CHUNK
```

```python
data = audio[start:end]
fig,ax = plt.subplots()
ax.plot(time[start:end],numpy.fft.fft(data))
start +=CHUNK
end+=CHUNK
data = numpy.sort(data)
print(data[-9:]*100)
```
